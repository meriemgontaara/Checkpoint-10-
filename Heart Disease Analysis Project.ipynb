## Data Description

The data contains 303 observatons, thirteen features and one target as below:

- age: The person's age in years

- sex: The person's sex (1 = male, 0 = female)

- cp: The chest pain experienced (Value 1: typical angina, Value 2: atypical angina, Value 3: non-anginal pain, Value 4: asymptomatic)

- trestbps: The person's resting blood pressure (mm Hg on admission to the hospital)

- chol: The person's cholesterol measurement in mg/dl

- fbs: The person's fasting blood sugar (> 120 mg/dl, 1 = true; 0 = false)

- restecg: Resting electrocardiographic measurement (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria)

- thalach: The person's maximum heart rate achieved

- exang: Exercise induced angina (1 = yes; 0 = no)

- oldpeak: ST depression induced by exercise relative to rest

- slope: the slope of the peak exercise ST segment (Value 1: upsloping, Value 2: flat, Value 3: downsloping)

- ca: The number of major vessels (0-3)

- thal: A blood disorder called thalassemia (3 = normal; 6 = fixed defect; 7 = reversable defect)

- target: Heart disease (0 = no, 1 = yes)

This implies that we must include all of the features, as we can only eliminate those features where the correlation of two or more features is very high.

We split our data into train and test set. We chose only 4 features as input variables for our model because we assessed them to be good predictors based on their moderate correlation with the target. We chose to keep 20% of our data for final testing of our model.

### Logistic Regression

### Apply Logistic Regression

#### Build our Logistic Model

Here, we take by default 0.5 as a threshold.

- If p >= 0.5, then the patient has a heart disease.

- If p < 0.5, then the patient do not have a heart disease.

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression() 
```

#### Fitting Training Data

```python
logreg.fit(x_train, y_train)
```

#### Testing Model's Performance

```python
y_pred  = logreg.predict(x_test)  
```

#### Performance of the Logistic Regression

```python
print("Accuracy={:.2f}".format(logreg.score(x_test, y_test)))
```

The model does a good job predicting whether patients have heart diseases or not, with 87% accuracy.

#### Finding the threshold that maximizes accuracy in Logistic Regression

```python
import numpy as np
from sklearn.metrics import balanced_accuracy_score
```

We first run through all the scores and store the accuracies in a list. 

```python
threshold = []
accuracy = []

for p in np.unique(logreg.predict_proba(x_train)[:,1]):
  threshold.append(p)
  y_pred = (logreg.predict_proba(x_train)[:,1] >= p).astype(int)
  accuracy.append(balanced_accuracy_score(y_train,y_pred))
```

Then, we plot them.

```python
plt.scatter(threshold,accuracy)
plt.xlabel("Threshold")
plt.ylabel("Balanced accuracy")
plt.show()
```

As you can see, we have a maximum slightly before 0.5. The correct value is:

```python
print("Threshold that maximizes accuracy: ",threshold[np.argmax(accuracy)])
```

So, this is the threshold we are looking for.

#### Changing the Default Threshold for Logistic Regression

```python
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
```

```python
x, y = make_classification(
    n_classes=2, class_sep=1.5, weights=[0.9, 0.1],
    n_features=20, n_samples=1000, random_state=10
)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

logreg = LogisticRegression(class_weight="balanced")
logreg.fit(x_train, y_train)
THRESHOLD =  0.47659308894839764
preds = np.where(logreg.predict_proba(x_test)[:,1] > THRESHOLD, 1, 0)

pd.DataFrame(data=[accuracy_score(y_test, preds)], index=["Accuracy"])
```

Changing the default threshold by a threshold that maximizes accuracy has indeed maximized accuracy metric.
In fact, the model does a very good job predicting whether patients have a heart disease or not, with 92.8% accuracy.

#### Confusion Matrix

Just like accuracy, a confusion matrix is another way to evaluate Logistic Regression performance with more precision.

```python
confusion_matrix = pd.crosstab(y_test, preds, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

Here, confusion matrix tells us that:

- 13 passengers who did not have a heart disease are wrongly predicted as having a heart disease (False Positives)

- 5 passengers who have a heart disease are wrongly predicted as not having a heart disease (False Negatives)

- 267 passengers who do not have a heart disease also predicted as not having a heart disease (True Positives)

- 18 passengers who have a heart disease are also predicted as having a heart disease (True Negatives)

False Negative errors could be extremely costly, as they might lead a patient to believe that they do not have a heart disease, when in reality they are suffering from it. The top-right cell contains the false positive results. Although False Positive errors are less dangerous than a false negative result, they should also be avoided as they could lead to additional financial burden on the health care system, or additional stress for the patient, as additional tests or treatment may have to be provided. 

- Recall:
    
Out of all the passengers that do not have a heart disease, how many did we get correct? Recall = TP/(TP+FN)= 98.16%
- Precision:
    
Out of all the passengers that we predicted to not have a heart disease, how many did we actually get right? Precision= TP/(TP+FP)= 95.3%

## Another Validation Matrix: ROC/AUC

AUC - ROC curve is a performance measurement for the classification problems at various threshold settings. ROC (Receiver Operating Characteristic) is a probability curve and AUC (Area Under the Curve) represents the degree or measure of separability. It tells how much the model is capable of distinguishing between classes. The higher the AUC, the better the model is at predicting 0 classes as 0 and 1 classes as 1. By analogy, the Higher the AUC, the better the model is at distinguishing between patients who heart disease free and those who suffer from heart disease.

The ROC curve summarizes all of the confusion matrices that each threshold produced and is plotted with True Positive Rate (TPR) which is the same thing as recall metric (or sensitivity) against the False Positive Rate (FPR) where TPR is on the y-axis and FPR is on the x-axis.

```python
import matplotlib.pyplot as plt
from matplotlib import pyplot
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
# generate a no skill prediction (majority class)
ns_probs = [0 for _ in range(len(y_test))]
# predict probabilities
lr_probs = model.predict_proba(x_test)
dt_probs = model.predict_proba(x_test1)
clf_probs = model.predict_proba(x_test2)
knn_probs = model.predict_proba(x_test3)
# keep probabilities for the positive outcome only
lr_probs = lr_probs[:, 1]
dt_probs = dt_probs[:, 1]
clf_probs = clf_probs[:, 1]
knn_probs = knn_probs[:, 1]
# calculate scores
ns_auc = roc_auc_score(y_test, ns_probs)
lr_auc = roc_auc_score(y_test, lr_probs)
dt_auc = roc_auc_score(y_test1, dt_probs)
clf_auc = roc_auc_score(y_test2, clf_probs)
knn_auc = roc_auc_score(y_test3, knn_probs)
# summarize scores
print('No Skill: ROC AUC=%.3f' % (ns_auc))
print('Logistic Regression: ROC AUC=%.3f' % (lr_auc))
print('Decision Tree: ROC AUC=%.3f' % (dt_auc))
print('Random Forest: ROC AUC=%.3f' % (clf_auc))
print('KNN: ROC AUC=%.3f' % (knn_auc))
# calculate roc curves
ns_fpr, ns_tpr, _ = roc_curve(y_test, ns_probs)
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)
dt_fpr, dt_tpr, _ = roc_curve(y_test1, dt_probs)
clf_fpr, clf_tpr, _ = roc_curve(y_test2, clf_probs)
knn_fpr, knn_tpr, _ = roc_curve(y_test3, knn_probs)
# plot the roc curve for the model
pyplot.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
pyplot.plot(lr_fpr, lr_tpr, marker='.', label='Logistic Regression')
pyplot.plot(dt_fpr, dt_tpr, marker='.', label='Decision Tree')
pyplot.plot(clf_fpr, clf_tpr, marker='.', label='Random Forest')
pyplot.plot(knn_fpr, knn_tpr, marker='.', label='KNN')
# axis labels
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')
# show the legend
pyplot.legend()
# show the plot
pyplot.show()
```

The best ROC-AUC_Score is the one belonging to Logistic Regression which is equal to 0.933 which denotes a very good classifier able to predict patients that are actually having a heart disease as having a heart disease and patients that do not actually have a heart disease as not having a heart disease.
